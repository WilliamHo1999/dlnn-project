```python
from src.data.MNIST import MNISTBoundingBoxDataset, get_mnist_bounding_box_data_loader
import matplotlib.pyplot as plt
from torchvision import transforms

dataset = get_mnist_bounding_box_data_loader('Data/', batch_size = 125)
val_dataset = dataset['val']
tti = transforms.ToPILImage()

for i in val_dataset:
    for j, img in enumerate(i['image']):
        plt.imshow(tti(img))
        box = i['bounding_box']
        x_min = box[0][j]
        y_min = box[1][j]
        x_max = box[2][j]
        y_max = box[3][j]
        plt.scatter(x_min, y_min, color = 'red')
        plt.scatter(x_max, y_max, color = 'red')
        plt.show()
    break
```
